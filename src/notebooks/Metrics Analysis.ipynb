```python
# Imports

import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
# import seaborn as sb
```

```python
# Retrieve the calculated metrics and remove first 3 columns

df = pd.read_csv('class.csv')
df.head()
df.iloc[:25, 3:].head()

metrics = df.iloc[:, 3:]
metrics.head()
print(len(df.columns))
```

```python
# Calculate correlations of the metrics

correlation_matrix = metrics.corr()
# correlation_matrix.head()

correlation_matrix.style.background_gradient(axis=None)
```

```python
mask = np.zeros_like(correlation_matrix, dtype=bool)
mask[np.triu_indices_from(mask)] = True
correlation_matrix[mask] = np.nan
(correlation_matrix
 .style
 .background_gradient(axis=None)
 .highlight_null(null_color='#f1f1f1')  # Color NaNs grey
 .set_precision(2))

# dfi.export(correlation_matrix[mask],"mytable.png")
```
